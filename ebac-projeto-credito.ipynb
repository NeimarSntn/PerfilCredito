**Query:** select
sum(case when limite_credito between 0 and 10000 then 1 else 0 end) as faixa_0_10000,
sum(case when limite_credito between 10001 and 20000 then 1 else 0 end) as faixa_10001_20000,
sum(case when limite_credito between 20001 and 30000 then 1 else 0 end) as faixa_20001_30000,
sum(case when limite_credito between 30001 and 40000 then 1 else 0 end) as faixa_30001_40000,
sum(case when limite_credito between 40001 and 50000 then 1 else 0 end) as faixa_40001_50000
from credito

![Faixa por limite de crédito](https://github.com/NeimarSntn/Imagens/blob/main/Faixa%20por%20limite%20de%20cr%C3%A9dito.jpg?raw=true)

O limite de crédito está concentrado na faixa até 10.000 com uma fatia de 70%.

**Query:** select max(valor_transacoes_12m) as maior_valor_gasto, avg(valor_transacoes_12m) as media_valor_gasto, min(valor_transacoes_12m) as min_valor_gasto, sexo 
from credito 
group by sexo

![Gastos por Sexo](https://github.com/NeimarSntn/Imagens/blob/main/Gastos%20Por%20Sexo.jpg?raw=true)

Por mais a que o sexo masculino tenha uma maior concentração nas rendas maiores os gastos estão semelhantes entre os sexos.

**Query:** select avg(qtd_produtos) as qts_produtos, avg(valor_transacoes_12m) as media_valor_transacoes, avg(limite_credito) as media_limite, sexo, salario_anual
from credito 
where salario_anual != 'na' 
group by sexo, salario_anual 
order by avg(valor_transacoes_12m) desc

Filtrando a média de produtos e transações conforme a renda apresenta valores similares, quanto a média de limite no sexo feminino com a renda menor que 40K possui uma renda superior ao sexo masculino. 

Os dados evidenciam uma menor renda para o sexo feminino mesmo possuindo educação similiar, ressaltando que a amostra possui uma base menor do sexo feminino. O limite se concentra em na faixa até 10K, os gastos estão semelhantes entre os sexos.
